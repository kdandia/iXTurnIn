For the titanic competition, the dataset has passenger information for every passenger that was aboard the titanic on its first (and last trip). 

The target variable is whether the passenger died or not when the cruise ship sank
You can download the competition data (and check the data dictionary) on [kaggle](https://www.kaggle.com/c/titanic/data).

You will use the training data (file `train.csv`) to train your classifier, and will create submissions for the `test.csv`. 

Basically, you have to submit a csv file on the shape:

```

894,1

...
```

Where the PassengerId are the ids of the passengers on the `test.csv` dataset and `Survived` is your model prediction about the passenger (0, die, 1 survives).

In order to submit a file you have to create a profile on the website. Then you can upload the submission using the Website or using the [kaggle api](https://github.com/Kaggle/kaggle-api)

````python
from IPython.display import Image
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm
import warnings
warnings.simplefilter("ignore")
%matplotlib inline

matplotlib.rcParams['figure.figsize'] = [12, 12]
````

````python
train = pd.read_csv("data/train.csv")
train.head()
````

````python
train.dtypes
````

Missing Values

````python
agemedian = train.Age.median()
train.Age = train.Age.fillna(agemedian)
train.head()
````

````python
target = train['Survived']
target.head()
````

female = 1
male = 0

````python
train = train.replace('male', 0)
train = train.replace('female', 1)
train.head()
````

````python
independent = train[['Pclass', 'Sex', 'Age', 'Fare', 'SibSp']]
independent.head()
````

````python
independent.dtypes
````

````python
#train model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(independent, target)
````

````python
test = pd.read_csv("data/test.csv")
test.head()
````

````python
test.Age = test.Age.fillna(agemedian)
````

````python
test = test.replace('male', 0)
test = test.replace('female', 1)
test.head()
````

````python
predindependent = test[['Pclass', 'Sex', 'Age', 'Fare', 'SibSp']]
````

````python
predindependent.dtypes
````

````python
predindependent.isna().any()
````

````python
faremean = train.Fare.mean()
predindependent.Fare = predindependent.Fare.fillna(faremean)
````

````python
predindependent.isna().any()
````

````python
predictions = model.predict(predindependent)
predictions
````

````python
test['Survived'] = predictions
test.head()
````

````python
results = test[['PassengerId', 'Survived']]
results.head()
````

````python
results.to_csv("data/titanic_results1.csv", index = False)
````
